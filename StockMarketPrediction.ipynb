```python
df_new = df[['date', 'open', 'high', 'low', 'close', 'adjclose', 'volume', 'ticker']]
df_new
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_new["ticker"] = le.fit_transform(df_new["ticker"])
```

```python
df_new
```

```python
df_new.describe()
```

```python
df_new.info()
```

```python
df_new['year'] = pd.to_datetime(df_new['date']).dt.year
df_new['month'] = pd.to_datetime(df_new['date']).dt.month
df_new['day'] = pd.to_datetime(df_new['date']).dt.day
```

```python
df_new
```

```python
np.unique(df_new["year"])
```

```python
target = df["close"]
```

```python
df_new.drop(columns = ["date","year","close"],inplace = True)
```

```python
df_new.drop(columns = ["close"],inplace = True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df_new,target,test_size = 0.3,random_state = 4)
```

```python
len(X_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr_y_pred = lr.predict(X_test)
```

```python
mse_lr = mean_squared_error(y_test,lr_y_pred)
```

```python
mse_lr
```

```python
from sklearn.ensemble import RandomForestRegressor
Rr = RandomForestRegressor()
Rr.fit(X_train,y_train)
Rr_y_pred = Rr.predict(X_test)
mse_Rr = mean_squared_error(y_test,Rr_y_pred)
mse_Rr
```

```python
from xgboost import XGBRegressor
xg = XGBRegressor()
xg.fit(X_train,y_train)
xg_y_pred = xg.predict(X_test)
mse_xg = mean_squared_error(y_test,xg_y_pred)
mse_xg
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=3)
gb.fit(X_train,y_train)
gb_y_pred = gb.predict(X_test)
mse_gb = mean_squared_error(y_test,gb_y_pred)
mse_gb
```
